```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
## 1. Matplotlib

Matplotlib is a Python library used for creating static and customizable visualizations. It provides full control over plot elements and is widely used in data analysis and scientific research.
```

```python
### Line Plot

**Description:**  
A line plot is used to visualize trends over continuous data.

**Use Case:**  
Used to show changes over time.
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
### Scatter Plot

**Description:**  
Scatter plots show the relationship between two numerical variables.

**Use Case:**  
Used to identify correlations.
```

```python
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
### Bar Chart

**Description:**  
Bar charts compare values across categories.

**Use Case:**  
Used for categorical comparisons.
```

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.show()
```

```python
## 2. Seaborn

Seaborn is a statistical data visualization library built on top of Matplotlib. It provides better default styling and simpler syntax for complex plots.
```

```python
### Histogram

**Description:**  
A histogram shows the distribution of data.

**Use Case:**  
Used to analyze data frequency and spread.
```

```python
data = [10, 12, 14, 15, 18, 20, 22, 25, 28]

sns.histplot(data, kde=True)
plt.title("Seaborn Histogram")
plt.show()
```

```python
### Box Plot

**Description:**  
A box plot displays data distribution and outliers.

**Use Case:**  
Used to detect outliers.
```

```python
sns.boxplot(data=data)
plt.title("Seaborn Box Plot")
plt.show()
```
